```python
def fread(header, start, nbytes, precision ):    
    if (precision == 'uint8'):
        size = nbytes
        data = int.from_bytes(header[start: start+size])
    if (precision == 'int16'):
        size = nbytes *2
        data = int.from_bytes(header[start: start+size])
    if (precision == 'uint32'):
        size = nbytes *4
        data = int.from_bytes(header[start: start+size])
    if (precision == "float32"):
        size = nbytes * 4 
        data = float.from_bytes(header[start: start+size])
    return data , start+size
```
